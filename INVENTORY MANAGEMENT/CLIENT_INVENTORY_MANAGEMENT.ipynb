# Time Series Stationarity
## A time series is stationary if the mean and variance are fixed between two equidistance points.
## Most the models like ARIMA and it's submodels need the stationary time series data.
## Ket's check if our data has some trend/seasonality or it is stationary. For stationarity check we can conduct augmented dickey fuller test.

# ADFULLER TEST:

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller_test=adfuller(week_data["spare_parts"],autolag="AIC")
print("ADF test statistics:",adfuller_test[0])
print("p_value:",adfuller_test[1])
print("number of lags used:",adfuller_test[2])
print("Number of observation taken for adfuller and critical values:",adfuller_test[3])
for i,j in adfuller_test[4].items():
    print("crititcal values at",i,"is",j)
```

```python
week_data.head()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(week_data);
```

# DIFFERENCING:

```python
week_diff1=week_data.diff(periods=1)
```

```python
week_diff1.head()
```

### Remove the null values:

```python
plt.figure(figsize=(15,4),facecolor="white")
plt.plot(week_diff)
```

### To check whether it is stationary or not after the differencing it

```python
week_diff
```

```python
adfuller_test2=adfuller(week_diff["spare_parts"],autolag="AIC")
print("ADF test statistics:",adfuller_test2[0])
print("p_value:",adfuller_test2[1])
print("number of lags used:",adfuller_test2[2])
print("Number of observation taken for adfuller and critical values:",adfuller_test2[3])
for l,k in adfuller_test2[4].items():
    print("crititcal values at",l,"is",k)
```

## Here the p_value is 9.640971371890529e-06 which is less than 0.05,now we can conclude the pattern is stationary.Now we can use any of the models like Auto Regression,ARIMA and SARIMA to predict the future needs.

# HOLTS WINTER:

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
hw_model=ExponentialSmoothing(train["spare_parts"],trend="add",seasonal="add",seasonal_periods=12).fit()
```

```python
hw_predict = hw_model.forecast(len(test))
```

```python
print(hw_predict)
```

```python
plt.figure(figsize=(15,4),facecolor="white")
plt.plot(train,label="training_Data")
plt.plot(test,color="green",label="actual_values")
plt.plot(hw_predict,color="red",label="predicted_values")
plt.title("Training value with actual and predicted values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,4),facecolor="white")
plt.plot(test,color="green",label="actual_values")
plt.plot(hw_predict,color="red",label="predicted_values")
plt.title("Actual and Predicted values")
plt.legend()
plt.show()
```

## EVALUATION METRICS:

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mae_error = mean_absolute_error(test['spare_parts'],hw_predict)
print("The mean absolute error for the holt winter method is ->",mae_error)
```

# HYPERPARAMETER TUNING:

```python
import itertools
p=d=q=range(0,2)
P=D=Q=range(0,3)
s=12
pdq=list(itertools.product(p,d,q))
seasonal_pdq=[(x[0],x[1],x[2],s) for x in itertools.product(P,D,Q)]
```

```python
print("length of pdq->",len(pdq))
```

```python
print("LENGTH OF seasonal_pdq->",len(seasonal_pdq))
```

```python
print("pdq->",pdq)
```

```python
print("seasonal_pdq->",seasonal_pdq)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
parameters=[]
aic_value=[]
for params in pdq:
    for seasonal_params in seasonal_pdq:
        try:
            hyper_sarima=SARIMAX(train,order=params,seasonal_order=seasonal_params)
            hyper_sarima_fit=hyper_sarima.fit()
            print("order:",params,"seasonal_order:",seasonal_params,"AIC_values",hyper_sarima_fit.aic)
            parameters.append((params,seasonal_params))
            aic_value.append(hyper_sarima_fit.aic)
        except:
            continue
```

```python
low_aic=aic_value.index(min(aic_value))
low_aic
```

```python
hypered_params=parameters[low_aic]
hyper_order=hypered_params[0]
hyper_seasonal=hypered_params[1]
print("order",hyper_order)
print("seasonal",hyper_seasonal)
```

```python
hyper_sarima=SARIMAX(train["spare_parts"],order=(0, 0, 1),seasonal_order=(2, 2, 0, 12))
```

```python
hyper_sarima_result=hyper_sarima.fit()
```

```python
print(hyper_sarima_result.summary())
```

```python
hyper_sarima_predict=hyper_sarima_result.forecast(steps=16)
```

```python
print(hyper_sarima_predict)
```

```python
plt.figure(figsize=(15,4),facecolor="white")
plt.plot(train,label="training_Data")
plt.plot(test,color="green",label="actual_values")
plt.plot(hyper_sarima_predict,color="red",label="predicted_values")
plt.title("SARIMA MODEL WITH RANDOM VALUE")
plt.legend()
plt.show()
```

```python
mae_error_sarima_hyper= mean_absolute_error(test['spare_parts'],hyper_sarima_predict)
print("The mean absolute error for the holt winter method is ->",mae_error_sarima_hyper)
```

```python
mse_error_sarima_hyper= mean_squared_error(test['spare_parts'],hyper_sarima_predict)
print("The mean absolute error for the holt winter method is ->",mse_error_sarima_hyper)
```
